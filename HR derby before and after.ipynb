```python
from pybaseball import batting_stats_range
import pandas as pd
```

# Seeing whether players performance increase or decrease after HR derby performance in last 3 derbys

```python
derbys = pd.read_excel(r"C:\Users\Owner\OneDrive\Desktop\exploratory data files\HR derby names.xlsx")
derbys
```

```python
derbys_2021 = list(derbys["HR_2021"])
derbys_2019 = list(derbys["HR_2019"])
derbys_2018 = list(derbys["HR_2018"])
```

```python
before_2021 = batting_stats_range("2021-04-01", "2021-07-12").query('Name in @derbys_2021')
after_2021 = batting_stats_range("2021-07-12", "2021-11-01").query('Name in @derbys_2021')
before_2021
```

```python
before_2019 = batting_stats_range("2019-04-01", "2019-07-08").query('Name in @derbys_2019')
after_2019 = batting_stats_range("2019-07-08", "2019-11-01").query('Name in @derbys_2019')
before_2019
```

```python
before_2018 = batting_stats_range("2018-04-01", "2018-07-16").query('Name in @derbys_2018')
after_2018 = batting_stats_range("2018-07-16", "2018-11-01").query('Name in @derbys_2018')
before_2018
```

```python
before_2018["B-A"] = "Before"
before_2019["B-A"] = "Before"
before_2021["B-A"] = "Before"
after_2018["B-A"] = "After"
after_2019["B-A"] = "After"
after_2021["B-A"] = "After"
```

```python
before_2018["K_rate"] = before_2018["SO"]/ before_2018["PA"]
before_2019["K_rate"] = before_2019["SO"]/ before_2019["PA"]
before_2021["K_rate"] = before_2021["SO"]/ before_2021["PA"]
after_2018["K_rate"] = after_2018["SO"]/ after_2018["PA"]
after_2019["K_rate"] = after_2019["SO"]/ after_2019["PA"]
after_2021["K_rate"] = after_2021["SO"]/ after_2021["PA"]
```

```python
before_2018["BB_rate"] = before_2018["BB"]/ before_2018["PA"]
before_2019["BB_rate"] = before_2019["BB"]/ before_2019["PA"]
before_2021["BB_rate"] = before_2021["BB"]/ before_2021["PA"]
after_2018["BB_rate"] = after_2018["BB"]/ after_2018["PA"]
after_2019["BB_rate"] = after_2019["BB"]/ after_2019["PA"]
after_2021["BB_rate"] = after_2021["BB"]/ after_2021["PA"]
```

```python
list(before_2018.columns)
```

```python
pd.concat([before_2018, after_2018]).sort_values("Name")[["Name", "B-A","R", "HR", "RBI", "BA", "OBP", "SLG", "OPS", "K_rate", "BB_rate"]]
```

```python
pd.concat([before_2019, after_2019]).sort_values("Name")[["Name", "B-A","R", "HR", "RBI", "BA", "OBP", "SLG", "OPS", "K_rate", "BB_rate"]]
```

```python
pd.concat([before_2021, after_2021]).sort_values("Name")[["Name", "B-A", "R", "HR", "RBI", "BA", "OBP", "SLG", "OPS", "K_rate", "BB_rate"]]
```

### Since the second half of the season players generally don't have the same amount of PA before and after- I will be focusing on BA, OBP, SLG, OPS, K + BB rate

## The idea that players get worse after the HR derby is not necessarily true; it is on a case by case basis
## Some players do get worse like for instance ohtani in 2021 but some players get better like soto in 2021 or Harper in 2018
## In general most players have similar SLG, OBP, OPS before and after the HR derby
## Since this is only shortly before the trade deadline it probably won't affect any trade value
